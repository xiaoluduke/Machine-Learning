## 1
Calculate the (a) first and (b) second derivative of $f(x) = x^3$. lastly, (c) numerically evaluate the second derivative for $x=3$.

**ANSWER**
The first and second derivatives are calculated as follows:

**(a)** The first derivative:
$$ \frac{df}{dx} = 3x^2$$

**(b)** The second derivative:
$$ \frac{d^2f}{dx^2} = 6x$$

```python
# (c) Numerically evaluate the second derivative for f(x)

# Initialize variables for the analysis
x = 3

# Compute the derivative
df2 = 6 * x

print('(c) The second derivative at x = 3 is {}'.format(df2))
```

### Figure Example:

First load some data to plot (the example here is from the lecture on bias-variance tradeoff):

```python
import numpy as np
import matplotlib.pyplot as plt
import pickle

# Load the data for plotting
datafilename = './plot_example_data/data.pkl'
infile = open(datafilename,'rb')
loaded_data = pickle.load(infile)
infile.close()

# Store the loaded data in convenient variable names
kvalues = loaded_data['kvalues']
error_training = loaded_data['error_training']
error_testing = loaded_data['error_testing']
error_bayesclf = loaded_data['error_bayesclf']

# Choose colors that will be distinguishable from one another and look professional
color0 = '#121619' # Dark grey
color1 = '#00B050' # Green
color2 = '#7c7c7c' # Light grey
```

Then create the plot:

```python
%config InlineBackend.figure_format = 'retina' # Makes the plots clear on high-res screens

# Create the plot
plt.figure(figsize=(7,5), dpi= 100) # Adjust the figure size and dpi (dots per inch) 
                                    # to clearly display the plot and balance plot proportions
plt.semilogx(kvalues,error_training,color=color0,label='Training (in-sample)')
plt.semilogx(kvalues,error_testing,color=color1,label='Test (out-of-sample)')
plt.semilogx(kvalues,error_bayesclf,'--',color=color2,label='Bayes (optimal)')
plt.legend()
plt.grid('on')
plt.xlabel('k nearest neighbors') # Always use X and Y labels - add units if appropriate as well
                                  # If there are color bars (there are not here) - label that axis as well
plt.ylabel('Binary Classification Error Rate')
plt.axis([1,200,0,0.5]) # Ensure the axis is the right size for the plot 
                        # and typically, you should start the y-axis at 0
plt.tight_layout() # Use this to maximize the use of space in the figure
plt.show()
```
